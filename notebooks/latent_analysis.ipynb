# Latent space analysis

Encode test set, run t-SNE on the means, and decode a 2D grid traversal
for a model trained with `latent_dim=2`.

```python
import torch
from src.model import VAE
from src.data import get_loaders
from src.visualize import encode_dataset, tsne_plot, latent_grid

model = VAE(latent_dim=20)
model.load_state_dict(torch.load('runs/default/last.pt', map_location='cpu'))
model.eval()
_, test_dl = get_loaders(batch_size=128)
mu, y = encode_dataset(model, test_dl, max_n=2000)
tsne_plot(mu, y, 'samples/tsne.png')
```

```python
# 2D grid (requires latent_dim=2 model)
m2 = VAE(latent_dim=2)
m2.load_state_dict(torch.load('runs/latent2d/last.pt', map_location='cpu'))
m2.eval()
g = latent_grid(m2, n=20)
import matplotlib.pyplot as plt
plt.figure(figsize=(8,8)); plt.imshow(g[0], cmap='gray'); plt.axis('off')
```
