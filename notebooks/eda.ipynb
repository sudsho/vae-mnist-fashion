# Fashion-MNIST EDA

Quick look at class balance, pixel value distributions, and a few sample images per class.

```python
import numpy as np
import matplotlib.pyplot as plt
from torchvision import datasets, transforms

ds = datasets.FashionMNIST('./data', train=True, download=True, transform=transforms.ToTensor())
len(ds)
```

```python
labels = np.array([y for _, y in ds])
np.bincount(labels)
```

```python
# pixel histogram
X = np.stack([ds[i][0].numpy() for i in range(2000)]).reshape(-1)
plt.hist(X, bins=50); plt.show()
```

```python
classes = ['t-shirt','trouser','pullover','dress','coat','sandal','shirt','sneaker','bag','boot']
fig, ax = plt.subplots(2, 5, figsize=(10, 4))
for c in range(10):
    idx = np.where(labels==c)[0][0]
    ax[c//5, c%5].imshow(ds[idx][0].squeeze(), cmap='gray')
    ax[c//5, c%5].set_title(classes[c]); ax[c//5, c%5].axis('off')
plt.tight_layout()
```
